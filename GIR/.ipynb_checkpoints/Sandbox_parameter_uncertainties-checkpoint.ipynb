```python
## writing the gas parameter sampling definition:

def get_more_gas_cycle_params(N,choose_gases=['CO2','CH4','N2O'],CH4_forc_feedbacks=False, help=False):
    
    param_defaults = get_gas_parameter_defaults(choose_gases=choose_gases,CH4_forc_feedbacks=CH4_forc_feedbacks)

    param_uncert = pd.read_pickle('./Parameter_Sets/Complete_parameter_uncertainty.p')

    param_ensemble = pd.concat(N*[param_defaults['default']],keys=np.arange(N),axis=1)

    for gas in choose_gases:

        for param in param_defaults.index:

            select_param = param_uncert.loc[param,('default',gas)]

            if select_param:

                param_sample = select_param[0].rvs(*select_param[1],N)

                param_ensemble.loc[param,(slice(None),gas)] = param_sample
    
    return param_ensemble
```

# Comparing thermal response to CMIP6 models

```python
## lets get some CMIP6 data:

import xarray as xr
import zarr
import gcsfs

xr.set_options(display_style='html')

df = pd.read_csv('https://storage.googleapis.com/cmip6/cmip6-zarr-consolidated-stores.csv')
df.head()
```

```python
def get_all_models_var_exp_global_av(var,exp):

    model_list = list(set(df.query("variable_id == 'areacella'").source_id.unique()).intersection(set(df.query("activity_id=='CMIP' & table_id == 'Amon' & variable_id == '"+var+"' & experiment_id == '"+exp+"'").source_id)))

    out_dict = {}

    for i,model in enumerate(model_list):

        clear_output(wait=True)
        display(var+'_'+exp+' model: '+model+' ('+str(i+1)+'/'+str(len(model_list))+')')

        df_ta = df.query("activity_id=='CMIP' & table_id == 'Amon' & variable_id == '"+var+"' & experiment_id == '"+exp+"' & source_id == '"+model+"'")

        gcs = gcsfs.GCSFileSystem(token='anon')

        # get the path to a specific zarr store (the first one from the dataframe above)
        zstore = df_ta.zstore.values[-1]

        # create a mutable-mapping-style interface to the store
        mapper = gcs.get_mapper(zstore)

        # open it using xarray and zarr
        ds = xr.open_zarr(mapper, consolidated=True)

        df_area = df.query("variable_id == 'areacella' & source_id == '"+model+"'")
        ds_area = xr.open_zarr(gcs.get_mapper(df_area.zstore.values[0]), consolidated=True)

        total_area = ds_area.areacella.sum(dim=[x for x in list(ds_area.dims) if x[0]=='l' and len(x) in [3,8,9]])
        ta_timeseries = ((ds[var] * ds_area.areacella).sum(dim=[x for x in list(ds_area.dims) if x[0]=='l' and len(x) in [3,8,9]]) / total_area)
        ta_timeseries.load()

        out_dict[model] = ta_timeseries[:].to_pandas()
        
    return out_dict
```

```python
abrupt_4xCO2_tas = get_all_models_var_exp_global_av('tas','abrupt-4xCO2')
picontrol_tas = get_all_models_var_exp_global_av('tas','piControl')
onepctCO2_tas = get_all_models_var_exp_global_av('tas','1pctCO2')
```

```python
def get_all_models_surface_var_exp_global_av(var,exp):

    model_list = list(set(df.query("variable_id == 'areacella'").source_id.unique()).intersection(set(df.query("activity_id=='CMIP' & table_id == 'Amon' & variable_id == '"+var+"' & experiment_id == '"+exp+"'").source_id)))

    out_dict = {}

    for i,model in enumerate(model_list):

        clear_output(wait=True)
        display(var+'_'+exp+' model: '+model+' ('+str(i+1)+'/'+str(len(model_list))+')')

        df_ta = df.query("activity_id=='CMIP' & table_id == 'Amon' & variable_id == '"+var+"' & experiment_id == '"+exp+"' & source_id == '"+model+"'")

        gcs = gcsfs.GCSFileSystem(token='anon')

        # get the path to a specific zarr store (the first one from the dataframe above)
        zstore = df_ta.zstore.values[-1]

        # create a mutable-mapping-style interface to the store
        mapper = gcs.get_mapper(zstore)

        # open it using xarray and zarr
        ds = xr.open_zarr(mapper, consolidated=True)

        df_area = df.query("variable_id == 'areacella' & source_id == '"+model+"'")
        ds_area = xr.open_zarr(gcs.get_mapper(df_area.zstore.values[0]), consolidated=True)

        total_area = ds_area.areacella.sum(dim=[x for x in list(ds_area.dims) if x[0]=='l' and len(x) in [3,8,9]])
        ta_timeseries = ((ds[var].sel(plev=max(ds['plev'].values)) * ds_area.areacella).sum(dim=[x for x in list(ds_area.dims) if x[0]=='l' and len(x) in [3,8,9]]) / total_area)
        ta_timeseries.load()

        out_dict[model] = ta_timeseries[:].to_pandas()
        
    return out_dict
```
